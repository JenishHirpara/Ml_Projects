```python
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
new_img = cv2.flip(fix_img, 0)
```

```python
plt.imshow(new_img)
```

```python
plt.imshow(fix_img)
```

```python
new_img = np.copy(fix_img)
```

```python
cv2.rectangle(new_img,pt1=(1500,900),pt2=(3700,4000),color=(255,0,0),thickness=30)
```

```python
plt.imshow(new_img)
```

```python
new_img_2 = np.copy(fix_img)
```

```python
vertices = np.array([ [2500,500], [1200,3800], [3800,3800] ], dtype=np.int32)
```

```python
pts = vertices.reshape((-1,1,2))
```

```python
cv2.polylines(new_img_2,[pts],isClosed=True,color=(255,0,0),thickness=30)
plt.imshow(new_img_2)
```

```python
cv2.fillPoly(new_img_2, [pts], color=(255,0,0))
```

```python
plt.imshow(new_img_2)
```

```python
new_img_3 = np.copy(fix_img)
```

```python
new_img_4 = cv2.resize(new_img_3,(800,500))
```

```python
#############
## FUNCTION ##
##############

def draw_circle(event,x,y,flags,param):
    
    if event == cv2.EVENT_RBUTTONDOWN:
        cv2.circle(new_img_4,(x,y),100,(0,0,255),thickness=30)

cv2.namedWindow(winname='my_drawing')

cv2.setMouseCallback('my_drawing', draw_circle)

####################
## SHOWING IMAGE WITH OPENCV ##
####################

while True:
    cv2.imshow('my_drawing', new_img_4)
    
    if cv2.waitKey(20) & 0XFF == 27:
        break
        
cv2.destroyAllWindows()
```
