```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('../numpy_images/sammy.jpg')
```

```python
plt.imshow(img)
```

```python
edges = cv2.Canny(image=img,threshold1=127,threshold2=127)
```

```python
plt.imshow(edges)
```

```python
edges = cv2.Canny(image=img,threshold1=0,threshold2=255)
```

```python
plt.imshow(edges)
```

```python
med_val = np.median(img)
```

```python
med_val
```

```python
# Lower threshold to either 0 or 70% of the median value whichever is greater
lower = int(max(0,0.7*med_val))
# Upper threshold to either 130% of the median or the max 255, whichever is smaller
upper = int(min(255,1.3*med_val))
```

```python
edges = cv2.Canny(image=img,threshold1=lower,threshold2=upper+50)
plt.imshow(edges)
```
