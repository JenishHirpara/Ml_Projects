```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display(img):
    fig = plt.figure(figsize=(12,10))
    ax=fig.add_subplot(111)
    ax.imshow(img)
```

```python
def load_img():
    img = cv2.imread('../numpy_images/a-pair-of-giraffes.jpg')
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    return img
```

```python
img = load_img()
display(img)
```
