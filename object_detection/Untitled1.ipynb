```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("../numpy_images/a-pair-of-giraffes.jpg", cv2.IMREAD_GRAYSCALE)
m,n = img.shape
L = img.max()
```

```python
plt.imshow(img, cmap="gray")
```

```python
img_negative = L-img
cv2.imwrite('grayscale input.png', img)
cv2.imwrite('Negative image.png', img_negative)
```

```python
plt.imshow(img_negative, cmap="gray")
```

```python
T=100
img_thresh = np.zeros((m,n), dtype = int)

for i in range(m):
    for j in range(n):
        if img[i,j] < T:
            img_thresh[i,j]= 0
        else:
            img_thresh[i,j] = L

cv2.imwrite('Thresh.png', img_thresh)
```

```python
plt.imshow(img_thresh, cmap="gray")
```

```python
T1=80
T2=200

img_thresh_back = np.zeros((m,n), dtype = int)

for i in range(m):
    for j in range(n):
        if T1 <img[i,j] < T2:
            img_thresh_back[i,j]= L
        else:
            img_thresh_back[i,j] = img[i,j]


cv2.imwrite('GreySlicing with BG.png', img_thresh_back)
```

```python
plt.imshow(img_thresh_back, cmap="gray")
```

```python
img_thresh_nobg = np.zeros((m,n), dtype = int)
for i in range(m):
    for j in range(n):
        if T1 <img[i,j] < T2:
            img_thresh_nobg[i,j]= L
        else:
            img_thresh_nobg[i,j] = 0

cv2.imwrite('GreySlicing without BG.png', img_thresh_nobg)
```

```python
plt.imshow(img_thresh_nobg, cmap="gray")
```
