```python
import cv2
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('../numpy_images/pexels-monstera-5063470.jpg')
```

```python
# MATPLOTLIB --> RGB
# OPENCV --> BGR
```

```python
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_img)
```

```python
img_gray = cv2.imread('../numpy_images/pexels-monstera-5063470.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img_gray, cmap='gray')
```

```python
plt.imshow(fix_img)
```

```python
fix_img.shape
```

```python
new_img = cv2.resize(fix_img, (4000, 2000))
```

```python
plt.imshow(new_img)
```

```python
w_ratio = 0.8
h_ratio = 0.2
```

```python
new_img = cv2.resize(fix_img, (0,0), fix_img, w_ratio, h_ratio)
```

```python
plt.imshow(new_img)
```

```python
new_img.shape
```

```python
new_img = cv2.flip(fix_img, 0)
plt.imshow(new_img)
```

```python
new_img = cv2.flip(fix_img, 1)
plt.imshow(new_img)
```

```python
new_img = cv2.flip(fix_img, -1)
plt.imshow(new_img)
```
